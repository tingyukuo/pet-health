## 斷詞

```python
# 讀檔：一條一條讀進來
Trainlist = []
for line in pet_health_raw['item']:
    Trainlist.append(line)
```

```python
# 斷詞
Train_seg=[]
for i in range(len(Trainlist)):
    Train_seg.append([' '.join(jieba.cut(Trainlist[i], cut_all=False))])
# 因為會跑很久，檢核是否資料有持續在跑
    if i % 100 == 0 :
        print(i)
```

```python
pet_health = pd.DataFrame(Train_seg, columns =['item'])
pet_health['class']=pet_health_raw['class']
pet_health.head()
```

## CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_transformer = CountVectorizer().fit(pet_health['item'])
```

```python
print(len(bow_transformer.vocabulary_))
```

```python
pet10 = pet_health['item'][9]
print(pet10)
```

```python
bow10 = bow_transformer.transform([pet10])
print(bow10)
print(bow10.shape)
```

```python
pet7 = pet_health['item'][6]
print(pet7)
```

```python
bow7 = bow_transformer.transform([pet7])
print(bow7)
print(bow7.shape)
```

```python
print(bow_transformer.get_feature_names_out()[742])
print(bow_transformer.get_feature_names_out()[978])
```

```python
pet_bow = bow_transformer.transform(pet_health['item'])
```

```python
print('Shape of Sparse Matrix: ', pet_bow.shape)
print('Amount of Non-Zero occurences: ', pet_bow.nnz)
```

```python
sparsity = (100.0 * pet_bow.nnz / (pet_bow.shape[0] * pet_bow.shape[1]))
print('sparsity: {}'.format(round(sparsity)))
```

## TfidfTransformer

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer = TfidfTransformer().fit(pet_bow)
tfidf7 = tfidf_transformer.transform(bow7)
print(tfidf7)
```

```python
#check what is the IDF (inverse document frequency) of the word
#print(tfidf_transformer.idf_[bow_transformer.vocabulary_['word']])
```

```python
pet_tfidf = tfidf_transformer.transform(pet_bow)
print(pet_tfidf.shape)
```

## Training a model - naive_bayes

```python
from sklearn.naive_bayes import MultinomialNB
health_detect_model = MultinomialNB().fit(pet_tfidf, pet_health['class'])
print('predicted:', health_detect_model.predict(tfidf7)[0])
print('expected:', pet_health['class'][6])
```

## Model Evaluation

```python
all_predictions = health_detect_model.predict(pet_tfidf)
print(all_predictions)
```

```python
from sklearn.metrics import classification_report
print (classification_report(pet_health['class'], all_predictions))
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

pet_train, pet_test, label_train, label_test = \
train_test_split(pet_health['item'], pet_health['class'], test_size=0.2)

print(len(pet_train), len(pet_test), len(pet_train) + len(pet_test))
```

## Data Pipeline

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('bow', CountVectorizer()),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', MultinomialNB()),  # train on TF-IDF vectors w/ Naive Bayes classifier
])
```

```python
pipeline.fit(pet_train,label_train)
```

```python
predictions_test = pipeline.predict(pet_test)
```

```python
predictions_test
```

```python
print(classification_report(predictions_test,label_test))
```

```python
pipeline.score(pet_test,label_test)
```

```python
#確認test data結果
pred_test = predictions_test.tolist()
pet_health_a = pd.DataFrame(pet_test)
pet_health_b = pd.DataFrame(label_test)
pet_health_chk = pet_health_a.join(pet_health_b)
pet_health_chk['pred_test'] = pred_test
#pet_health_chk
pet_health_chk.to_csv('pet_health_chk.csv')
```
